```python
# Importamos la data necesaria para el proceso
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importamos las primeras 1000000 lineas de 2000.csv
vuelos_2005 = pd.read_csv('2003.csv', nrows=100000)
```

```python
vuelos_2005.head()
```

```python
# CAMBIANDO LOS NOMBRES DE TODAS LAS COLUMNAS

# cambiar nombre de la columna 'Year' a 'Año'
vuelos_2005.rename(columns={'Year':'Anio'}, inplace=True)

# cambiar nombre de la columna 'Month' a 'Mes'
vuelos_2005.rename(columns={'Month':'Mes'}, inplace=True)

# cambiar nombre de la columna 'DayofMonth' a 'Dia'
vuelos_2005.rename(columns={'DayofMonth':'Dia'}, inplace=True)

# cambiar nombre de la columna 'DayOfWeek' a 'Dia de la semana'
vuelos_2005.rename(columns={'DayOfWeek':'Dia_Sem'}, inplace=True)

# cambiar nombre de la columna 'UniqueCarrier' a 'Aerolinea'
vuelos_2005.rename(columns={'UniqueCarrier':'ID_Aerolinea'}, inplace=True)

# cambiar nombre de la columna 'Origin' a 'ID_Aeropuerto_O'
vuelos_2005.rename(columns={'Origin':'ID_Aeropuerto_O'}, inplace=True)

# cambiar nombre de la columna 'Dest' a 'ID_Aeropuerto_D'
vuelos_2005.rename(columns={'Dest':'ID_Aeropuerto_D'}, inplace=True)

# cambiar nombre de la columna 'DepTime' a 'Hora_SalidaR'
vuelos_2005.rename(columns={'DepTime':'Hora_SalidaR'}, inplace=True)

# cambiar nombre de la columna 'CRSDepTime' a 'Hora_SalidaP'
vuelos_2005.rename(columns={'CRSDepTime':'Hora_SalidaP'}, inplace=True)

# cambiar nombre de la columna 'ArrTime' a 'Hora_LlegadaR'
vuelos_2005.rename(columns={'ArrTime':'Hora_LlegadaR'}, inplace=True)

# cambiar nombre de la columna 'CRSArrTime' a 'Hora_LlegadaP'
vuelos_2005.rename(columns={'CRSArrTime':'Hora_LlegadaP'}, inplace=True)

# cambiar nombre de la columna 'CRSElapsedTime' a 'Duracion_Promedio'
vuelos_2005.rename(columns={'CRSElapsedTime':'Duracion_Promedio'}, inplace=True)

# cambiar nombre de la columna 'ArrDelay' a 'Retraso_Llegada'
vuelos_2005.rename(columns={'ArrDelay':'Retraso_Llegada'}, inplace=True)

# cambiar nombre de la columna 'DepDelay' a 'Retraso_Salida'
vuelos_2005.rename(columns={'DepDelay':'Retraso_Salida'}, inplace=True)

# cambiar nombre de la columna 'Distance' a 'Distancia'
vuelos_2005.rename(columns={'Distance':'Distancia'}, inplace=True)

# cambiar nombre de la columna 'Cancelled' a 'Cancelado'
vuelos_2005.rename(columns={'Cancelled':'Cancelado'}, inplace=True)

# cambiar nombre de la columna 'CancellationCode' a 'Codigo_Cancelacion'
vuelos_2005.rename(columns={'CancellationCode':'Codigo_Cancelacion'}, inplace=True)

# cambiar nombre de la columna 'CarrierDelay' a 'Retraso_Aerolinea'
vuelos_2005.rename(columns={'CarrierDelay':'Retraso_Aerolinea'}, inplace=True)

# cambiar nombre de la columna 'WeatherDelay' a 'Retraso_Clima'
vuelos_2005.rename(columns={'WeatherDelay':'Retraso_Clima'}, inplace=True)

# cambiar nombre de la columna 'NASDelay' a 'Retraso_NAS'
vuelos_2005.rename(columns={'NASDelay':'Retraso_NAS'}, inplace=True)

# cambiar nombre de la columna 'SecurityDelay' a 'Retraso_Seguridad'
vuelos_2005.rename(columns={'SecurityDelay':'Retraso_Seguridad'}, inplace=True)

# cambiar nombre de la columna 'LateAircraftDelay' a 'Retraso_Aeronave'
vuelos_2005.rename(columns={'LateAircraftDelay':'Retraso_Aeronave'}, inplace=True)

# Reordenamos las columnas

vuelos_2005 = vuelos_2005[['Anio', 'Mes', 'Dia', 'Dia_Sem', 'ID_Aerolinea', 'ID_Aeropuerto_O', 'ID_Aeropuerto_D', 
'Hora_SalidaR', 'Hora_SalidaP', 'Hora_LlegadaR', 'Hora_LlegadaP', 'Duracion_Promedio', 'Retraso_Llegada', 'Retraso_Salida', 
'Distancia', 'Cancelado', 'Codigo_Cancelacion', 'Retraso_Aerolinea', 'Retraso_Clima', 'Retraso_NAS', 'Retraso_Seguridad', 
'Retraso_Aeronave']]
```

```python
import numpy as np

# Crear una columna 'Estado_Vuelo', que tome valores 'Codigo_Cancelacion'
vuelos_2005['ID_EstadoVuelo'] = np.where(vuelos_2005['Codigo_Cancelacion'].notnull(), vuelos_2005['Codigo_Cancelacion'], np.where(vuelos_2005['Cancelado'] == 1, 'ND', 'E'))

# Eliminar columna 'Codigo_Cancelacion'
vuelos_2005.drop('Codigo_Cancelacion', axis=1, inplace=True)
vuelos_2005.drop('Cancelado', axis=1, inplace=True)

# Eliminar las columnas de retrasos
vuelos_2005.drop('Retraso_Aerolinea', axis=1, inplace=True)
vuelos_2005.drop('Retraso_NAS', axis=1, inplace=True)
vuelos_2005.drop('Retraso_Seguridad', axis=1, inplace=True)
vuelos_2005.drop('Retraso_Aeronave', axis=1, inplace=True) 
vuelos_2005.drop('Retraso_Clima', axis=1, inplace=True)


# Creando una columna de 'Partida_A_Tiempo', A: 'Retraso_Salida' <= 0, B: 0 < 'Retraso_Salida' <= 15, C: 'Retraso_Salida' > 15, NE: 'Retraso_Salida' = NaN
vuelos_2005['ID_DepTime'] = np.where(vuelos_2005['Retraso_Salida'] <= 0, 'A', np.where(vuelos_2005['Retraso_Salida'] <= 15, 'B', np.where(vuelos_2005['Retraso_Salida'] > 15, 'C', 'ND')))

# Creando una columna de 'Llegada_A_Tiempo', A: 'Retraso_Llegada' <= 0, B: 0 < 'Retraso_Llegada' <= 15, C: 'Retraso_Llegada' > 15, NE: 'Retraso_Llegada' = NaN
vuelos_2005['ID_ArrTime'] = np.where(vuelos_2005['Retraso_Llegada'] <= 0, 'A', np.where(vuelos_2005['Retraso_Llegada'] <= 15, 'B', np.where(vuelos_2005['Retraso_Llegada'] > 15, 'C', 'ND')))

# Eliminar las columnas de retrasos
vuelos_2005.drop('Retraso_Llegada', axis=1, inplace=True)
vuelos_2005.drop('Retraso_Salida', axis=1, inplace=True)
```
