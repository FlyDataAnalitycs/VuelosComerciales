```python
# Importamos la data necesaria para el proceso
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importamos las primeras 1000000 lineas de 2000.csv
vuelos_2008 = pd.read_csv('2008.csv', nrows=100000)
```

```python
vuelos_2008.head()
```

```python
# CAMBIANDO LOS NOMBRES DE TODAS LAS COLUMNAS

vuelos_2008.rename(columns={'Year':'Anio'}, inplace=True)
vuelos_2008.rename(columns={'Month':'Mes'}, inplace=True)
vuelos_2008.rename(columns={'DayofMonth':'Dia'}, inplace=True)
vuelos_2008.rename(columns={'DayOfWeek':'Dia_Sem'}, inplace=True)
vuelos_2008.rename(columns={'UniqueCarrier':'ID_Aerolinea'}, inplace=True)
vuelos_2008.rename(columns={'Origin':'ID_Aeropuerto_O'}, inplace=True)
vuelos_2008.rename(columns={'Dest':'ID_Aeropuerto_D'}, inplace=True)
vuelos_2008.rename(columns={'DepTime':'Hora_SalidaR'}, inplace=True)
vuelos_2008.rename(columns={'CRSDepTime':'Hora_SalidaP'}, inplace=True)
vuelos_2008.rename(columns={'ArrTime':'Hora_LlegadaR'}, inplace=True)
vuelos_2008.rename(columns={'CRSArrTime':'Hora_LlegadaP'}, inplace=True)
vuelos_2008.rename(columns={'CRSElapsedTime':'Duracion_Promedio'}, inplace=True)
vuelos_2008.rename(columns={'ArrDelay':'Retraso_Llegada'}, inplace=True)
vuelos_2008.rename(columns={'DepDelay':'Retraso_Salida'}, inplace=True)
vuelos_2008.rename(columns={'Distance':'Distancia'}, inplace=True)
vuelos_2008.rename(columns={'Cancelled':'Cancelado'}, inplace=True)
vuelos_2008.rename(columns={'CancellationCode':'Codigo_Cancelacion'}, inplace=True)
vuelos_2008.rename(columns={'CarrierDelay':'Retraso_Aerolinea'}, inplace=True)
vuelos_2008.rename(columns={'WeatherDelay':'Retraso_Clima'}, inplace=True)
vuelos_2008.rename(columns={'NASDelay':'Retraso_NAS'}, inplace=True)
vuelos_2008.rename(columns={'SecurityDelay':'Retraso_Seguridad'}, inplace=True)
vuelos_2008.rename(columns={'LateAircraftDelay':'Retraso_Aeronave'}, inplace=True)

# Reordenamos las columnas

vuelos_2008 = vuelos_2008[['Anio', 'Mes', 'Dia', 'Dia_Sem', 'ID_Aerolinea', 'ID_Aeropuerto_O', 'ID_Aeropuerto_D', 
'Hora_SalidaR', 'Hora_SalidaP', 'Hora_LlegadaR', 'Hora_LlegadaP', 'Duracion_Promedio', 'Retraso_Llegada', 'Retraso_Salida', 
'Distancia', 'Cancelado', 'Codigo_Cancelacion', 'Retraso_Aerolinea', 'Retraso_Clima', 'Retraso_NAS', 'Retraso_Seguridad', 
'Retraso_Aeronave']]
```

```python
import numpy as np

# Creando una columna 'Estado_Vuelo', que tome valores 'Codigo_Cancelacion'
vuelos_2008['ID_EstadoVuelo'] = np.where(vuelos_2008['Codigo_Cancelacion'].notnull(), vuelos_2008['Codigo_Cancelacion'], np.where(vuelos_2008['Cancelado'] == 1, 'ND', 'E'))

# Eliminando columna 'Codigo_Cancelacion'
vuelos_2008.drop('Codigo_Cancelacion', axis=1, inplace=True)
vuelos_2008.drop('Cancelado', axis=1, inplace=True)

# Eliminando las columnas de retrasos
vuelos_2008.drop('Retraso_Aerolinea', axis=1, inplace=True)
vuelos_2008.drop('Retraso_NAS', axis=1, inplace=True)
vuelos_2008.drop('Retraso_Seguridad', axis=1, inplace=True)
vuelos_2008.drop('Retraso_Aeronave', axis=1, inplace=True) 
vuelos_2008.drop('Retraso_Clima', axis=1, inplace=True)


# Creando una columna de 'Partida_A_Tiempo', A: 'Retraso_Salida' <= 0, B: 0 < 'Retraso_Salida' <= 15, C: 'Retraso_Salida' > 15, NE: 'Retraso_Salida' = NaN
vuelos_2008['ID_DepTime'] = np.where(vuelos_2008['Retraso_Salida'] <= 0, 'A', np.where(vuelos_2008['Retraso_Salida'] <= 15, 'B', np.where(vuelos_2008['Retraso_Salida'] > 15, 'C', 'ND')))

# Creando una columna de 'Llegada_A_Tiempo', A: 'Retraso_Llegada' <= 0, B: 0 < 'Retraso_Llegada' <= 15, C: 'Retraso_Llegada' > 15, NE: 'Retraso_Llegada' = NaN
vuelos_2008['ID_ArrTime'] = np.where(vuelos_2008['Retraso_Llegada'] <= 0, 'A', np.where(vuelos_2008['Retraso_Llegada'] <= 15, 'B', np.where(vuelos_2008['Retraso_Llegada'] > 15, 'C', 'ND')))

# Eliminando las columnas de retrasos
vuelos_2008.drop('Retraso_Llegada', axis=1, inplace=True)
vuelos_2008.drop('Retraso_Salida', axis=1, inplace=True)
```
