### concatenando la data

```python
import numpy as np
import pandas as pd
import math as math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv('clima_exitosos2.csv')
data2 = pd.read_csv('clima_cancelados2.csv')

# combinando data1 y data2 en orden de ID_Vuelo
data = pd.concat([data1, data2], axis=0, ignore_index=True)
```

```python
data.info()
```

### Analisis de variables a considerar

```python
# Haciendo un grafico de dispersion de las variables
sns.pairplot(data, hue='bool', vars=['temperatura','precipitacion','direccion_viento',
                                    'velocidad_viento','presion','cobertura_nubes'])
```

```python
# Mostrar la matriz de correlacin
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(), annot=True, fmt='.1f')
plt.show()
```

Las unicas columnas significativas son cobertura_nubes y temperatura

```python
# Elimanaminando las columnas de precipitacion, direccion_viento, velocidad_viento y presion
data = data.drop(['precipitacion','direccion_viento','velocidad_viento','presion'], axis=1)

# Exportando la data a un archivo csv
data.to_csv('data.csv', index=False)
```
